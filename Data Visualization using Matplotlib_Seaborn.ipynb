```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file. Also,'df' is data frame.
df = pd.read_csv(r"C:\Users\EliteBook 840 G3\OneDrive\Customers.csv")
df
```

```python
df.shape
```

```python
# Customers' information
df.dtypes
```

```python
df.columns
```

```python
#Dropping some irrelevant columns
df.head(1)
```

```python
df.drop(["Prefix", "FirstName", "LastName","Unnamed: 5", "Unnamed: 6", "Unnamed: 8", "Unnamed: 17", "Unnamed: 18", "BirthDate.1" ], axis=1, inplace=True)
```

```python
df.shape
```

```python
df.head()
# you can also say 'df.columns' to display the relevant columns left after dropping the irrelevant columns.
```

```python
#Check for missing values
df.isna().sum()
```

```python
# Filling the 'Gender' column with specific members and not generalizing the fill.
df['Gender'].bfill(inplace = True)
df.isna().sum()
```

```python
#renaming the "Unnamed: 4" column to Fullname
df.rename(columns={"Unnamed: 4":"Fullname"}, inplace=True)
df.head()
```

```python
#'str.strip()' is used to completely remove a particular character without replacing it. The '$' sign in AnnualIncome makes it a string, so we need to remove it.
df.AnnualIncome = df.AnnualIncome .str.strip("$")
df.head()
```

```python
# 'str.replace()' is used to replace a particular character with another character or nothing.
df.AnnualIncome = df.AnnualIncome.str.replace(',', '')
df.head()
```

```python
#astype is for type casting, used to convert one datatype of a column to another; changing our currency to an integer
df.AnnualIncome = df.AnnualIncome.astype(int)
```

```python
#Converting the BirthDate column to Datetime datatype
df["BirthDate"] = pd.to_datetime(df["BirthDate"], format ='mixed')
df.sample(4)
```

```python
df['MaritalStatus'].replace({'M':'Married','S': 'Single',}, inplace = True)
df['Gender'].replace({'M':'Male','F': 'Female',}, inplace = True)
df['HomeOwner'].replace({'N':'No','Y': 'Yes',}, inplace = True)
df.sample(5)
```

# Data Visualization

## 1. Creating a Bar Chart Showing the count of customers by MaritalStatus

```python
MS_bar = df.groupby("MaritalStatus")["MaritalStatus"].count()
#plt.figure(figsize=(6, 4))
plt.bar(MS_bar.index, MS_bar, color="blue")
plt.xlabel("Marital Status")
plt.ylabel("Total Customer")
plt.title("Count of Customers by Marital Status")
plt.show()
```

## 12. Pie chart showing the proportion of different Marital Status categories within a specific Occupation

```python
# Get unique occupations from the DataFrame
unique_occupations = df['Occupation'].unique()
unique_occupations
```

```python
# Plot pie charts for each occupation
for occupation in unique_occupations:
    # Filter the DataFrame for the specific occupation
    filtered_data = df[df['Occupation'] == occupation]
    
    # Count the occurrences of each MaritalStatus category with the occupation 
    status_counts = filtered_data['MaritalStatus'].value_counts()
    
    # Plot the pie chart
    plt.figure(figsize=(4, 2))
    plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
    
    # Add title
    plt.title('Proportion of MaritalStatus categories within {}'.format(occupation))
    
    # Display the plot
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

## 17. Pie chart to represent the proportion of customers by Occupation within a specific income range

```python
occupation_counts = df['Occupation'].value_counts()
occupation_counts
```

```python
df.head(2)
```

```python
# Get unique values of AnnualIncome
unique_incomes = df['AnnualIncome'].unique()

# Find minimum and maximum annual income
min_income = df['AnnualIncome'].min()
max_income = df['AnnualIncome'].max()

print("Unique Annual Income Values:", unique_incomes)
print("Minimum Annual Income:", min_income)
print("Maximum Annual Income:", max_income)
```

```python
# Filter data based on income range
filtered_df = df[(df['AnnualIncome'] >= min_income) & (df['AnnualIncome'] <= max_income)]

# Calculate proportion of customers by occupation
occupation_counts = filtered_df['Occupation'].value_counts(normalize=True)

# Plot pie chart
plt.figure(figsize=(8, 8))
occupation_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Customers by Occupation within Income Range $10,000 - $170,000')
plt.ylabel('')  # No need to label the y-axis
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
